## **Data Cleaning**


```python
# Convert release_date to datetime
df["release_date"] = pd.to_datetime(df["release_date"], errors="coerce")

# Fill missing numeric values with median (common EDA approach)
df["runtime"] = df["runtime"].fillna(df["runtime"].median())

# Fill missing text fields with 'Unknown'
text_cols = ["genres","keywords","cast","crew","director","overview","tagline","homepage"]
for c in text_cols:
    if c in df.columns:
        df[c] = df[c].fillna("Unknown")

# Drop duplicate rows (if any)
df = df.drop_duplicates()

df.isna().sum().sort_values(ascending=False).head(10)

```

---


## **Univariate Analysis**


### Runtime distribution


```python
sns.histplot(df["runtime"], bins=20, kde=True)
plt.title("Runtime Distribution")
plt.xlabel("Runtime (minutes)")
plt.show()

```

### Revenue distribution


```python
sns.boxplot(x=df["revenue"])
plt.title("Revenue Distribution")
plt.show()

```

### Vote average distribution


```python
sns.histplot(df["vote_average"], bins=20, kde=True)
plt.title("Vote Average Distribution")
plt.xlabel("vote_average")
plt.show()

```

---


## **Bivariate Analysis**


### Average revenue by language (Top 10 languages by count)


```python
top_lang = df["original_language"].value_counts().head(10).index
avg_rev_by_lang = df[df["original_language"].isin(top_lang)].groupby("original_language")["revenue"].mean().sort_values(ascending=False)
avg_rev_by_lang.plot(kind="bar")
plt.title("Average Revenue by Original Language (Top 10 languages)")
plt.ylabel("Average Revenue")
plt.show()

```

### Rating by status


```python
sns.barplot(x="status", y="vote_average", data=df, estimator=np.mean)
plt.title("Average Vote Average by Status")
plt.xticks(rotation=45)
plt.show()

```

### Budget vs Revenue


```python
plt.scatter(df["budget"], df["revenue"], alpha=0.5)
plt.title("Budget vs Revenue")
plt.xlabel("Budget")
plt.ylabel("Revenue")
plt.show()

```

---


## **Multivariate Analysis**


### Revenue by Status & Language (Top 5 languages)


```python
top5_lang = df["original_language"].value_counts().head(5).index
df_small = df[df["original_language"].isin(top5_lang)]

sns.catplot(x="status", y="revenue", hue="original_language", kind="bar", data=df_small, height=5, aspect=1.6)
plt.xticks(rotation=45)
plt.show()

```

## **Correlation Matrix**


```python
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Matrix")
plt.show()

```
